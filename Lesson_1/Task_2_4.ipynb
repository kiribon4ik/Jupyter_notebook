Задание 4

Создайте датафрейм authors_stat на основе информации из authors_price. В датафрейме authors_stat должны быть четыре столбца:
author_name, min_price, max_price и mean_price,
в которых должны содержаться соответственно имя автора,минимальная, максимальная и средняя цена на книги этого автора.

```python
import numpy as np
import pandas as pd
```

```python
authors = pd.DataFrame({'author_id':[1, 2, 3],
                        'author_name':['Тургенев', 
                                       'Чехов', 
                                       'Островский']},
                        columns=['author_id', 'author_name'])
authors.set_index('author_id', inplace=True)
authors
```

```python
books = pd.DataFrame({'author_id':[1, 1, 1, 2, 2, 3, 3],
                     'book_title':['Отцы и дети', 
                                   'Рудин', 
                                   'Дворянское гнездо', 
                                   'Толстый и тонкий', 
                                   'Дама с собачкой', 
                                   'Гроза', 
                                   'Таланты и поклонники'],
                     'price':[450, 300, 350, 500, 450, 370, 290]},
                     columns=['author_id', 'book_title', 'price'])
books.set_index('author_id', inplace=True)
books
```

```python
authors_price = pd.merge(authors, books, on = 'author_id', how='outer')
authors_price
```

```python
authors_stat = pd.concat([authors_price.groupby('author_name').agg({'price':'min'}),
                        authors_price.groupby('author_name').agg({'price':'max'}),
                        authors_price.groupby('author_name').agg({'price':'mean'})], axis=1, ignore_index=True)

authors_stat = authors_stat.rename(columns={0:'min_price', 
                                            1:'max_price', 
                                            2:'mean_price'})

authors_stat
```
